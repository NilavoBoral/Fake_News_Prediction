# Preprocessing

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re     # regular expression
```

```python
nltk.download('stopwords')
```

```python
ps = PorterStemmer()
```

```python
nilavo = []
for i in x['title']:
    i = re.sub('[^a-zA-Z]',' ',i)
    i = i.lower().split()
    i = [ps.stem(j) for j in i if not j in stopwords.words('english')]
    i = ' '.join(i)
    nilavo.append(i)
```

```python
nilavo
```

```python
import tensorflow
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocabulary_size = 10000
onehot = [one_hot(i, vocabulary_size) for i in nilavo]
onehot
```

```python
onehot_pad = pad_sequences(onehot, padding='pre', maxlen=20)
onehot_pad
```

```python
onehot_pad[0]
```

```python
len(onehot_pad[0])
```

# Creating Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
features = 40
def build(hp):
    model = Sequential()
    model.add(Embedding(vocabulary_size, features, input_length= 20))
    model.add(LSTM(hp.Int('neurons', min_value=32, max_value=512, step=16)))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])), metrics=['accuracy'])
    return model
```

```python
from kerastuner import RandomSearch
```

```python
tuner_search = RandomSearch(
    build,
    objective='val_accuracy',
    max_trials=5,
    directory='output',
    project_name="Fake news"
)
```

# Model training

```python
onehot_pad.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(np.array(onehot_pad), np.array(y), test_size= 0.3, random_state= 42)
```

```python
tuner_search.search(x_train, y_train, epochs=3, validation_split= 0.1)
```

```python
model=tuner_search.get_best_models(num_models=1)[0]
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=64)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
t = []
for i in y_pred:
    if i < 0.5:
        i=0
    else:
        i=1
    t.append(i)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, t)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, t)
```
